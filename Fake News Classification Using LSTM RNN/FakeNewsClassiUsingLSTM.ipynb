```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
### vocabulary size
voc_size = 5000
```

## One Hot Representation

```python
messages = X.copy()
```

```python
messages.head(10)
```

```python
messages['title'][5]
```

```python
messages.reset_index(inplace = True)
```

```python
messages.head(10)
```

```python
messages['title'][6]
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
set_stopwords = set(stopwords.words('english'))
```

```python
## Data Preprocessing
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []

for i in range(len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if word not in set_stopwords]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

# One Hot Embedding

```python
one_hot_repr = [one_hot(word, voc_size) for word in corpus]
one_hot_repr
```

## Embedding Representation

```python
sent_length = 20
embedded_docs = pad_sequences(one_hot_repr, padding='post', maxlen = sent_length)
```

```python
embedded_docs
```

```python
## creating model
embedding_feature_lenth = 40
model = Sequential()
model.add(Embedding(voc_size, embedding_feature_lenth, input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
len(embedded_docs), y.shape
```

```python
import numpy as np

X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
X_final.shape, y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

## Model Training

```python
## Finally Trainning
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, batch_size=64)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = np.where(y_pred > 0.5, 1, 0) ## AUC ROC Curve
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

## Adding Dropout

```python
from tensorflow.keras.layers import Dropout
```

```python
## creating model with Dropout
embedding_feature_lenth = 40
model = Sequential()
model.add(Embedding(voc_size, embedding_feature_lenth, input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
import numpy as np

X_final = np.array(embedded_docs)
y_final = np.array(y)
```
